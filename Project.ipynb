### 20. Classify Movies Based on Ratings [Excellent, Good, and Average]

```python
new_project.columns
```

```python
# We are creating our own functions to classify Movies based on Ratings.

# def rating - function name
# (rating) - let me pass argument

def rating(rating):
    if rating >= 7.0:
       return "Excellent"
    elif rating >= 6.0:
       return "Good"
    else:
       return "Average"
```

```python
new_project['rating_cat']=new_project['Rating'].apply(rating)
```

### 22. Find Unique Values From Genre

```python
new_project.Genre
```

```python
# we are going to perform three steps to find unique values in Genre column
# 1. Split items with commas
# 2. Convert this 2D list to 1D list to find unique values
# 3. Find unique value in this empty list
```

```python
list1=[]
for value in new_project['Genre']:
    list1.append(value.split(','))

# to create one list for all items
# here we split items by comma. Then append this value over empty list
```

```python
list1
```

```python
# let me create this empty list for one_d
# 2. In this second step, 2D list is converted to 1D list.
one_d=[]
for item in list1:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
# third step, find unique value in this empty list

uni_list=[]
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)
```

```python
uni_list
```
